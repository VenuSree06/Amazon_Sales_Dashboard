### 4.How are product ratings distributed?

```python
colors = ["#4F46E5", "#06B6D4", "#10B981"]

plt.figure(figsize=(8,4))

n, bins, patches = plt.hist(
    df["rating"].dropna(),
    bins=10,
    edgecolor="black"
)

for i, patch in enumerate(patches):
    patch.set_facecolor(colors[i % len(colors)])

plt.title(
    "Distribution of Product Ratings",
    fontsize=14,
    fontweight="bold"
)

plt.xlabel("Rating")
plt.ylabel("Number of Products")

plt.show()
```



### 6.Is there a relationship between product price and rating?

This analysis investigates whether product price influences customer ratings.

```python
df[["actual_price", "rating"]].isnull().sum()
```

```python
price_rating_df = df[
    ["actual_price", "rating"]
].dropna()

price_rating_df["rating_group"] = (
    price_rating_df["rating"]
    .round(0)
)
```

```python
df[["actual_price", "rating"]].dtypes
```

```python
df["actual_price"] = (
    df["actual_price"]
    .astype(str)
    .str.replace("₹", "", regex=False)
    .str.replace(",", "", regex=False)
)

df["actual_price"] = pd.to_numeric(
    df["actual_price"],
    errors="coerce"
)
```

```python
df[["actual_price", "rating"]].dtypes
```

```python
price_rating_df = df[
    ["actual_price", "rating"]
].dropna()

price_rating_df["rating_group"] = (
    price_rating_df["rating"]
    .round(0)
)

avg_price_rating = (
    price_rating_df
    .groupby("rating_group")["actual_price"]
    .mean()
)

avg_price_rating.plot(
    kind="bar",
    figsize=(8,4),
    color=["#4F46E5", "#06B6D4", "#10B981"] * 3
)

plt.title(
    "Average Price by Rating Group",
    fontsize=14,
    fontweight="bold"
)

plt.xlabel("Rating Group")
plt.ylabel("Average Price (₹)")

plt.show()
```

### 7.Which product categories offer the highest average discount percentage?

This analysis identifies categories that receive the highest discounts.

```python
df["discount_percentage"] = (
    df["discount_percentage"]
    .astype(str)
    .str.replace("%", "", regex=False)
)

df["discount_percentage"] = pd.to_numeric(
    df["discount_percentage"],
    errors="coerce"
)
```

```python
category_discount = (
    df.groupby("category")["discount_percentage"]
    .mean()
    .sort_values(ascending=False)
    .head(10)
)

category_discount
```

```python
colors = ["#4F46E5", "#06B6D4", "#10B981"]

ax = category_discount.plot(
    kind="barh",
    figsize=(8,4),
    color=colors * 4
)

for container in ax.containers:
    ax.bar_label(container, fmt="%.1f")

plt.title(
    "Top 10 Categories by Average Discount Percentage",
    fontsize=22,
    fontweight="bold"
)

plt.xlabel("Average Discount (%)")
plt.ylabel("Category")

plt.show()
```

# Seaborn Visualizations

Seaborn provides advanced statistical visualizations built on top of Matplotlib.

The following charts provide deeper insights into Amazon product data.

## Seaborn Chart 1
### Distribution of Product Ratings

```python
plt.figure(figsize=(8,6))

sns.histplot(
    df["rating"].dropna(),
    bins=10,
    kde=True
)

plt.title(
    "Distribution of Product Ratings",
    fontsize=14,
    fontweight="bold"
)

plt.xlabel("Rating")
plt.ylabel("Count")

plt.show()
```

## Seaborn Chart 2

### Price Distribution by Rating Group

```python
price_rating_df = df[
    ["actual_price", "rating"]
].dropna()

price_rating_df["rating_group"] = (
    price_rating_df["rating"]
    .round(0)
)

plt.figure(figsize=(8,4))

sns.boxplot(
    x="rating_group",
    y="actual_price",
    data=price_rating_df
)

plt.title(
    "Price Distribution by Rating Group",
    fontsize=14,
    fontweight="bold"
)

plt.xlabel("Rating Group")
plt.ylabel("Price (₹)")

plt.show()
```

## Seaborn Chart 3

### Top Categories by Average Discount Percentage


```python
top_discount_categories = (
    df.groupby("category")["discount_percentage"]
    .mean()
    .sort_values(ascending=False)
    .head(10)
)

top_discount_categories
```

```python
plt.figure(figsize=(8,4))

sns.barplot(
    x=top_discount_categories.values,
    y=top_discount_categories.index,
)

plt.title(
    "Top 10 Categories by Average Discount",
    fontsize=20,
    fontweight="bold"
)

plt.xlabel("Average Discount (%)")
plt.ylabel("Category")

plt.show()
```
